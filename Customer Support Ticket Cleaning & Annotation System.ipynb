```
sample_tickets = [
    {
        "ticket_id": "TKT001",
        "subject": "Printer not working",
        "description": "My printter is not working. It keeps showing an error code 404. I NEED HELP ASAP. recieved it last week."
    },
    {
        "ticket_id": "TKT002",
        "subject": "Login issues",
        "description": "Hello, hope you're well. My login credentials (user@example.com) arE not working. Can you plesae reset my password? Thanks for your help!"
    },
    {
        "ticket_id": "TKT003",
        "subject": "Software Bug",
        "description": "The new softwaRe version 2.1.3 has a bug. It crasheS randomly. This affects Project Phoenix. FIX IT!!!"
    },
    {
        "ticket_id": "TKT004",
        "subject": "Account access",
        "description": "I cannOt access my account since 2023-10-26. My username is JohnDoe_789. Please check what's wrong. Best regards."
    },
    {
        "ticket_id": "TKT005",
        "subject": "Order #12345 delayed",
        "description": "My order number is #12345. It was supposed to arrive on Nov 1st, but its still pending. WHat is the status??!!"
    },
    {
        "ticket_id": "TKT006",
        "subject": "Email not sending",
        "description": "Emails are not sendiNG from my outlook. I've triEd restarting the app but it dosent work. Is there a server issue? Contact me at 555-123-4567."
    }
]

print(f"Generated {len(sample_tickets)} sample tickets.")
```

```
import re
import string

cleaned_descriptions = []

for ticket in sample_tickets:
    description = ticket['description']

    description = description.lower()

    description = description.translate(str.maketrans('', '', string.punctuation))

    description = re.sub(r'\d+', '', description)

    description = re.sub(r'\s+', ' ', description).strip()

    cleaned_descriptions.append(description)

print("Original vs. Cleaned Descriptions (First 3 examples):")
for i in range(min(3, len(sample_tickets))):
    print(f"\nOriginal: {sample_tickets[i]['description']}")
    print(f"Cleaned:  {cleaned_descriptions[i]}")
```

```
from textblob import TextBlob

spell_corrected_descriptions = []
irrelevant_phrases = [
    "hello hope youre well",
    "thanks for your help",
    "best regards",
    "i need help asap",
    "fix it",
    "what is the status",
    "contact me at"
]

for description in cleaned_descriptions:
    corrected_text = str(TextBlob(description).correct())

    for phrase in irrelevant_phrases:
        corrected_text = re.sub(re.escape(phrase), '', corrected_text, flags=re.IGNORECASE).strip()

    corrected_text = re.sub(r'\s+', ' ', corrected_text).strip()

    spell_corrected_descriptions.append(corrected_text)

print("Original Cleaned vs. Spell Corrected and Irrelevant Content Removed (First 3 examples):")
for i in range(min(3, len(cleaned_descriptions))):
    print(f"\nOriginal Cleaned: {cleaned_descriptions[i]}")
    print(f"Corrected:        {spell_corrected_descriptions[i]}")
```

```
import spacy

nlp = spacy.load("en_core_web_sm")

processed_tickets_spacy = []

for description in spell_corrected_descriptions:
    doc = nlp(description)

    processed_tokens = [token.lemma_ for token in doc if not token.is_punct and not token.is_stop and token.is_alpha]

    processed_text = " ".join(processed_tokens)
    processed_tickets_spacy.append(processed_text)

print("Spell Corrected vs. SpaCy Processed (First 3 examples):")
for i in range(min(3, len(spell_corrected_descriptions))):
    print(f"\nOriginal Corrected: {spell_corrected_descriptions[i]}")
    print(f"SpaCy Processed:  {processed_tickets_spacy[i]}")
```

```
named_entities_list = []

for description in spell_corrected_descriptions:
    doc = nlp(description)
    entities_for_ticket = []
    for ent in doc.ents:
        entities_for_ticket.append({
            'text': ent.text,
            'label': ent.label_,
            'start': ent.start_char,
            'end': ent.end_char
        })
    named_entities_list.append(entities_for_ticket)

print("Spell Corrected Descriptions with Extracted Named Entities (First 3 examples):")
for i in range(min(3, len(spell_corrected_descriptions))):
    print(f"\nOriginal Corrected: {spell_corrected_descriptions[i]}")
    print(f"Named Entities:     {named_entities_list[i]}")
```

```
print("\n--- Pipeline Effectiveness Comparison ---\n")

num_examples = 3

for i in range(min(num_examples, len(sample_tickets))):
    print(f"--- Ticket ID: {sample_tickets[i]['ticket_id']} ---")
    print(f"Original Description:                  {sample_tickets[i]['description']}")
    print(f"Initial Cleaned:                       {cleaned_descriptions[i]}")
    print(f"Spell Corrected & Irrelevant Removed:  {spell_corrected_descriptions[i]}")
    print(f"SpaCy Processed (Tokens/Lemmas):       {processed_tickets_spacy[i]}")
    print(f"Extracted Entities:                    {named_entities_list[i]}")
    print("\n")
```

```
import pandas as pd

df_tickets = pd.read_csv('/content/aa_dataset-tickets-multi-lang-5-2-50-version.csv')
df_tickets = df_tickets.head(5)

print(f"Loaded {len(df_tickets)} rows into df_tickets.")
print(df_tickets.head())
```

```
cleaned_descriptions_initial = []

for description in df_tickets['body']:
    # Convert to lowercase
    description = description.lower()

    # Remove punctuation
    description = description.translate(str.maketrans('', '', string.punctuation))

    # Remove numbers
    description = re.sub(r'\d+', '', description)

    # Remove extra whitespace and strip leading/trailing spaces
    description = re.sub(r'\s+', ' ', description).strip()

    cleaned_descriptions_initial.append(description)

print("Original vs. Cleaned Descriptions (First 3 examples):")
for i in range(min(3, len(df_tickets))):
    print(f"\nOriginal: {df_tickets['body'].iloc[i]}")
    print(f"Cleaned:  {cleaned_descriptions_initial[i]}")
```

```
from textblob import TextBlob

spell_corrected_descriptions_new = []
irrelevant_phrases = [
    "dear customer support team",
    "dear support team",
    "i hope this message reaches you well",
    "i am writing to report",
    "could you please provide",
    "i hope this message finds you well",
    "thank you for your inquiry",
    "hello hope youre well",
    "thanks for your help",
    "best regards",
    "i need help asap",
    "fix it",
    "what is the status",
    "contact me at"
]

for description in cleaned_descriptions_initial:
    corrected_text = str(TextBlob(description).correct())

    for phrase in irrelevant_phrases:
        corrected_text = re.sub(re.escape(phrase), '', corrected_text, flags=re.IGNORECASE).strip()

    corrected_text = re.sub(r'\s+', ' ', corrected_text).strip()

    spell_corrected_descriptions_new.append(corrected_text)

print("Original Cleaned vs. Spell Corrected and Irrelevant Content Removed (First 3 examples):")
for i in range(min(3, len(cleaned_descriptions_initial))):
    print(f"\nOriginal Cleaned: {cleaned_descriptions_initial[i]}")
    print(f"Corrected:        {spell_corrected_descriptions_new[i]}")
```

```
import spacy

nlp = spacy.load("en_core_web_sm")

processed_tickets_spacy_new = []

for description in spell_corrected_descriptions_new:
    doc = nlp(description)

    # Process tokens: lemma, remove punctuation, stop words, and non-alpha characters
    processed_tokens = [token.lemma_ for token in doc if not token.is_punct and not token.is_stop and token.is_alpha]

    processed_text = " ".join(processed_tokens)
    processed_tickets_spacy_new.append(processed_text)

print("Spell Corrected vs. SpaCy Processed (First 3 examples):")
for i in range(min(3, len(spell_corrected_descriptions_new))):
    print(f"\nOriginal Corrected: {spell_corrected_descriptions_new[i]}")
    print(f"SpaCy Processed:  {processed_tickets_spacy_new[i]}")
```

```
named_entities_list_new = []

for description in spell_corrected_descriptions_new:
    doc = nlp(description)
    entities_for_ticket = []
    for ent in doc.ents:
        entities_for_ticket.append({
            'text': ent.text,
            'label': ent.label_,
            'start': ent.start_char,
            'end': ent.end_char
        })
    named_entities_list_new.append(entities_for_ticket)

print("Spell Corrected Descriptions with Extracted Named Entities (First 3 examples):")
for i in range(min(3, len(spell_corrected_descriptions_new))):
    print(f"\nOriginal Corrected: {spell_corrected_descriptions_new[i]}")
    print(f"Named Entities:     {named_entities_list_new[i]}")
```

```
print("\n--- Pipeline Effectiveness Comparison ---\n")

num_examples = 3

for i in range(min(num_examples, len(df_tickets))):
    print(f"--- Ticket Index: {i+1} ---")
    print(f"Original Description:                  {df_tickets['body'].iloc[i]}")
    print(f"Initial Cleaned:                       {cleaned_descriptions_initial[i]}")
    print(f"Spell Corrected & Irrelevant Removed:  {spell_corrected_descriptions_new[i]}")
    print(f"SpaCy Processed (Tokens/Lemmas):       {processed_tickets_spacy_new[i]}")
    print(f"Extracted Entities:                    {named_entities_list_new[i]}")
    print("\n")
```
